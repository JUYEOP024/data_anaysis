```python
import matplotlib.pyplot as plt

months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
avg_temperatures = [-1, 0, 5, 10, 15, 20, 25, 23, 19, 12, 5, 0]
rainfall = [50, 40, 60, 80, 100, 120, 150, 140, 110, 90, 70, 60]
math_scores = [85, 90, 88, 76, 95, 89, 72, 84, 91, 87]
english_scores = [78, 82, 85, 80, 88, 90, 84, 79, 85, 83]
```

월별 평균 기온을 시각화하는 선 그래프를 그려보세요.

```python
plt.plot(months, avg_temperatures, marker='o', linestyle='-', color='b')
plt.title('Temperature avg months')
plt.xlabel('months')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

월별 강수량을 시각화하는 막대 그래프를 그려보세요.

```python
plt.bar(months, rainfall, color='skyblue')
plt.title('Months rainfall')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.show()
```

수학과 영어 점수의 분포를 비교하는 박스플롯 그래프를 그려보세요.

```python
data = [math_scores, english_scores]
plt.boxplot(data, labels=['math_scores','english_scores'])
plt.title('Math & English Scores')
plt.ylabel('Score')
plt.show()
```

수학 점수 분포를 분석하는 히스토그램을 그려보세요.

```python
plt.hist(math_scores, bins=10, color='green', edgecolor='black')
plt.title('Math_score')
plt.xlabel('score')
plt.ylabel('count')
plt.show()
```

수학 점수와 영어 점수의 상관관계를 분석하는 산점도를 그려보세요.

```python
plt.scatter(math_scores, english_scores, color='purple', alpha=0.6)
plt.title('Relationship between Study Hours and Exam Scores')
plt.xlabel('math_scores')
plt.ylabel('english_scores')
plt.show()
```
