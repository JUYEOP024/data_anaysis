## [실습 1] 다양한 그래프 활용 1

- 주어진 데이터에 대해 선 그래프, 막대 그래프, 히스토그램, 산점도를 한꺼번에 그려보세요.

```python
import matplotlib.pyplot as plt
import numpy as np



np.random.seed(0)

# 데이터 준비
x = np.arange(1, 11)
y = np.random.randint(10, 100, 10)
data = np.random.normal(50, 15, 100)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(2, 2, 1)
plt.plot(x,y,marker= 'o')

plt.subplot(2, 2, 2)
plt.bar(x,y,color='skyblue')
plt.subplot(2, 2, 3)
plt.hist(data,bins=20,color='orange',edgecolor = 'black')
plt.subplot(2, 2, 4)
plt.scatter(x,y,color = 'green',alpha = 0.3)
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
avg_temperatures = [-1, 0, 5, 10, 15, 20, 25, 23, 19, 12, 5, 0]
rainfall = [50, 40, 60, 80, 100, 120, 150, 140, 110, 90, 70, 60]
math_scores = [85, 90, 88, 76, 95, 89, 72, 84, 91, 87]|
english_scores = [78, 82, 85, 80, 88, 90, 84, 79, 85, 83]
```

월별 평균 기온을 시각화하는 선 그래프를 그려보세요.

```python
plt.plot(months,avg_temperatures,marker = 'o',linestyle='-',color = 'green')
```

월별 강수량을 시각화하는 막대 그래프를 그려보세요.

```python
plt.bar(months,rainfall,color='skyblue')
```

수학과 영어 점수의 분포를 비교하는 박스플롯 그래프를 그려보세요.

```python
data = [math_scores,english_scores]

plt.boxplot(data,labels=['Math','English'])
```

수학 점수 분포를 분석하는 히스토그램을 그려보세요.

```python
plt.hist(math_scores, bins=10, color='green', edgecolor='black')
```

수학 점수와 영어 점수의 상관관계를 분석하는 산점도를 그려보세요.

```python
plt.scatter(math_scores,english_scores,color='orange',alpha = 0.5)
```
