```python
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터셋 로드
titanic = sns.load_dataset('titanic')
```

- 문제 1: 생존자와 사망자 수 시각화
    - Titanic 데이터셋의 `survived` 열을 사용하여 생존자와 사망자의 수를 막대 그래프로 시각화하세요.
    - 그래프에 제목, 축 레이블을 추가하고 색상을 설정하세요.

```python
sns.countplot(x='survived', data=titanic,palette="pastel")
plt.title('survived')
plt.xlabel('1=yes,0=no')
plt.ylabel('인원수')
plt.show()
```

- 문제 2: 성별에 따른 생존율 비교
    - Titanic 데이터셋의 `sex`와 `survived` 열을 사용하여 성별에 따른 생존율을 막대 그래프로 시각화하세요.
    - 성별 생존율을 비교하고 제목, 축 레이블을 추가하세요

```python
sns.barplot(x='sex',y="survived", data=titanic,palette= "pastel")
plt.title('성별 생존율')
plt.xlabel('성별')
plt.ylabel('생존율')
plt.show()
```

- 문제 3: 클래스와 성별에 따른 생존율 비교
    - Titanic 데이터셋의 `class`, `sex`, `survived` 열을 사용하여 그룹별 생존율을 시각화하세요.
    - `class`를 x축으로 설정하고, 성별에 따라 색상을 구분하여 그래프를 생성하세요.
    - 그래프에 제목, 축 레이블, 범례를 추가하세요.

```python
sns.barplot(x="class",y="survived", hue='sex' ,data= titanic ,palette= "pastel")
plt.title('그룹별 생존율')
plt.xlabel('그룹')
plt.ylabel('생존자 수')
plt.show()
```

- 문제 4: 나이 분포 분석
    - Titanic 데이터셋의 `age` 열을 사용하여 승객들의 나이 분포를 히스토그램과 KDE 그래프로 시각화하세요.
    - 데이터에서 결측치를 제거하고 그래프를 그리세요.
    - 히스토그램의 막대 개수(`bins`)를 적절히 설정하고 제목, 축 레이블을 추가하세요.

```python
sns.histplot(titanic['age'].dropna(),
             kde = True,
             bins=30,
             color = "purple" 
) 
plt.title('나이분포도')
plt.xlabel('나이')
plt.ylabel('분포')
plt.show()
```

- 문제 5: 나이와 요금의 관계 분석
    - Titanic 데이터셋의 `age`, `fare`, `survived` 열을 사용하여 나이와 요금 간의 관계를 산점도로 시각화하세요.
    - 생존 여부(`survived`)에 따라 색상을 다르게 설정하세요.
    - 그래프에 제목, 축 레이블을 추가하세요.

```python
sns.scatterplot(x='age', y='fare',hue='survived',data=titanic,palette="pastel")

plt.title('나이와 요금의 관계')
plt.xlabel('나이')
plt.ylabel('요금')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 생성 (회사의 월별 제품 판매량)
data = {
    'January': [500, 600, 700, 400],
    'February': [550, 620, 680, 450],
    'March': [530, 610, 670, 420],
    'April': [580, 640, 710, 460],
    'May': [590, 660, 720, 470],
    'June': [620, 680, 750, 490]
}
companies = ['Company A', 'Company B', 'Company C', 'Company D']

df = pd.DataFrame(data, index = companies)
df
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df,annot = True, cmap ='YlGnBu',linewidths=0.5)

plt.title('MOnthly Product Sales by Company')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 수정된 데이터 (다양한 상관관계 유도)
data = {
    'Stock Price': [120, 135, 150, 160, 145, 170, 185, 190],    # 주가
    'Revenue': [50, 55, 65, 70, 67, 75, 80, 85],                # 매출
    'Profit': [30, 28, 25, 22, 21, 20, 18, 15],                # 이익 (감소하는 값)
    'Expenses': [60, 58, 55, 53, 50, 48, 45, 43],              # 비용 (감소)
    'Debt': [80, 75, 70, 68, 72, 74, 78, 82],                  # 부채 (변동)
    'R&D Spending': [10, 10, 15, 15, 20, 20, 25, 25]           # 연구개발비 (일정한 패턴)
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 상관계수 계산
corr_matrix = df.corr()

# Heatmap 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix,
             annot=True, 
             cmap='coolwarm', 
             linewidths=0.5)
# 타이틀 추가
plt.title('Correlation Between Financial Metrics')

# 그래프 보여주기
plt.show()

# 상관계수 출력
corr_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 예시 데이터 생성
data = {
    'Class': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Score': [85, 88, 90, 70, 75, 80],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}

# Violin Plot 그리기
plt.figure(figsize=(8, 6))
sns.violinplot(x='Class', y='Score', hue='Gender', data=data, split=True)

# 타이틀 추가
plt.title('Distribution of Exam Scores by Gender and Class')

# 그래프 보여주기
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# 아이리스 데이터셋 로드
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Pair Plot 그리기
sns.pairplot(df_iris, hue='species')

# 그래프 보여주기
plt.show()

```
