```python
import matplotlib.pyplot as plt
import numpy as np

# 2D 배열 데이터 생성
data = np.random.rand(10, 10)

# 두 개의 서브플롯에서 다른 컬러맵 적용
plt.subplot(1, 2, 1)
plt.imshow(data, cmap='plasma')
plt.title('Plasma Colormap')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.imshow(data, cmap='inferno')
plt.title('Inferno Colormap')
plt.colorbar()

# 서브플롯 간격 조정
plt.tight_layout()

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# 사용자 정의 컬러맵 설정
# LinearSegmentedColormap.from_list()를 사용하여 색상 리스트를 기반으로 컬러맵 정의
colors = ['#0000FF', "#000000", '#FF0000']  # 파란색에서 녹색, 빨간색으로 변환
custom_cmap = LinearSegmentedColormap.from_list('custom', colors)

# 2D 배열 데이터 생성
data = np.random.rand(10, 10)

# 사용자 정의 컬러맵 적용
plt.imshow(data, cmap=custom_cmap)
plt.colorbar()

# 타이틀 추가
plt.title('Custom Colormap')

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)#재현성 확보

x = np.arange(1,11)
y = np.random.randint(10,100,10)
data = np.random.normal(50,15,100)
plt.figure(figsize =(12, 10))

# 선 그래프 
plt.subplot(2, 2, 1)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Temperature Variation Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.grid(True)


#막대 그래프
plt.subplot(2, 2, 2)
plt.bar(x, y, color='skyblue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount')

#히스토그램
plt.subplot(2, 2, 3)
plt.hist(data, bins=10, color='green', edgecolor='black')
plt.title('Height Distribution of Students')
plt.xlabel('Height (cm)')
plt.ylabel('Number of Students')

#산점도
plt.subplot(2, 2, 4)
plt.scatter(x, y, color='purple', alpha=0.6)
plt.title('Relationship between Study Hours and Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')


plt.tight_layout()
plt.show() 
```

```python
import matplotlib.pyplot as plt

# 시장 점유율 데이터
labels = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [40, 30, 20, 10]

# Pie Chart 그리기
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral', 'gold'])

# 타이틀 추가
plt.title('Market Share by Company')

# 그래프 보여주기
plt.show()
```

```python
import matplotlib.pyplot as plt

# 시장 점유율 데이터
labels = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [40, 30, 20, 10]

# Donut Chart 그리기
plt.pie(
    market_share, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['skyblue', 'lightgreen', 'lightcoral', 'gold'], 
    wedgeprops={'width':0.5})

# 중앙에 텍스트 추가
plt.text(0, 0, 'Market Share', ha='center', va='center', fontsize=12)

# 타이틀 추가
plt.title('Donut Chart')

# 그래프 보여주기
plt.show()
```
